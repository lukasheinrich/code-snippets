```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D

test_data = np.array(
      [[[ 0.,  1],
        [ 0.,  2]],

       [[ 0.,  0],
        [ 0.,  1]]])
alphas = np.divide(test_data,np.max(test_data))/2.0
```

```python
colors = np.empty(test_data.shape + (4,))
colors[:,:,:] = [1,0,0,1.0]
colors[:,:,:,3] = alphas




fig = plt.figure(figsize=(10,10))
ax = fig.add_axes([0.1, 0.1, 0.7, 0.8], projection='3d')
_=ax.voxels(test_data, edgecolor=(0,0,0,0.0), facecolors=colors)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_aspect('equal')
ax.view_init(azim=45, elev=45)
```
